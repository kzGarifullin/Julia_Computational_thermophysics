```python
point =[ 
[0.5533604588440113, 3.127639293902885],

[0.5810593841117715, 2.6359713555404904],

[0.6034019418227745, 2.348842423917215],

[0.645932220727021, 1.9634284599437037],

[0.6841222346386084, 1.7270712823905725],

[0.7375562186377881, 1.496027728718478],

[0.830912270782805, 1.242951871957033],]
```

```python
def P_r(V_r, T_r):
    return 8*T_r/(3*V_r - 1) - 3 / V_r / V_r



v = [0.01*i for i in range(400)]
plt.figure(figsize=(15, 10))
T_range = [0.85, 0.88, 0.9, 0.93, 0.95, 0.97, 0.99]
i=0
for T in T_range:
    plt.plot(v[50:], np.array(P_r(np.array(v[50:]), T)), label=T)
    plt.plot(point[i][0], P_r(point[i][0], T), marker='o' , color='blue')
    plt.plot(point[i][1], P_r(point[i][1], T), marker='o', color='red')
    i+=1
plt.xlabel(r'$V$', fontsize=14)
plt.ylabel(r'$P$', fontsize=14)
#plt.plot(x, np.cos(x), label=r'$f_2(x)=\cos(x)$')
#plt.plot(x, -x, label=r'$f_3(x)=-x$')

plt.grid(True)

plt.legend(loc='best', fontsize=12)
plt.show()
```
