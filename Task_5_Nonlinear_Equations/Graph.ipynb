```python
def F(G, z, K):
    l = len(z)
    F = 0
    for i in range(l):
        F = F + z[i] * (K[i] - 1) / (G*(K[i] - 1) + 1)
    
    
    return F
```

```python
z = [0.2463, 0.2208, 0.2208, 0.3121]
K = [40, 25, 0.6, 0.005]

```

```python
import numpy as np
import matplotlib.pyplot as plt
x = [0.01*i for i in range(100)]
plt.figure(figsize=(10, 5))
plt.plot(x, np.array(F(np.array(x), z , K)), label=r'$F(G)$')
plt.plot(0.5506554018599481, 0, marker='o')
#plt.plot(x, np.cos(x), label=r'$f_2(x)=\cos(x)$')
#plt.plot(x, -x, label=r'$f_3(x)=-x$')
plt.xlabel(r'$G$', fontsize=14)
plt.ylabel(r'$F$', fontsize=14)
plt.grid(True)

plt.legend(loc='best', fontsize=12)
plt.show()
```
